#นักประเมินคนที่2 เป็นหลัก

```python
post2['ประเมิน1'] = post2['post id'].map(post1.set_index('post id')['ประเมินราคาซื้อ-ขาย']);
post2
```

```python
post2_1 = post2.rename(columns={"ประเมินราคาซื้อ-ขาย": "ประเมิน2"})
post2_1
```

```python
est2_1=post2_1[['post id','ราคาทรัพย์','ประเมิน1','ประเมิน2']]
est2_1
```

```python
est2_1.isnull().any()
```

```python
est2_1[['ประเมิน2']]  = est2_1[['ประเมิน2']].fillna(0)
```

```python
est2_1.isnull().any()
```

```python
est2_1
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
x_true = est2_1['ราคาทรัพย์']
y_pred = est2_1['ประเมิน1']
z_pred = est2_1['ประเมิน2']
```

```python
mean_absolute_percentage_error(x_true, y_pred)
print(f'ราคาตั้งขายเปรียบเทียบกับราคาประเมินทรัพย์ใช้ MAPE เป็นเกณฑ์ของนักประเมินคนที่ 1 เท่ากับ{mean_absolute_percentage_error(x_true, y_pred)}')
```

```python
mean_absolute_percentage_error(x_true, z_pred)
print(f'ราคาตั้งขายเปรียบเทียบกับราคาประเมินทรัพย์ใช้ MAPE เป็นเกณฑ์ของนักประเมินคนที่ 2 เท่ากับ{mean_absolute_percentage_error(x_true, z_pred)}')
```
